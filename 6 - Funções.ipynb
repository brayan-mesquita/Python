##contar até

```python
#contar de 0 até o numero que for digitado
def contar_Ate(numero):
  for i in range(0, numero+1):
    print(i)
contar_Ate(10)
```

##imprimir frase x vezes

```python
def imprirmir_x(frase, vezes):
  retorno = (frase + ' ') * vezes
  return retorno
imprirmir_x("Brayan", 100)
```

```python
def dividir(n, d=0):
  try: 
    print(int(n/d))
  except:
    print('O numero nao pode ser dividido por 0')
dividir(10)
dividir(10, 5)
```

```python
import random
secretNumber = random.randint(1, 20)
print('Adivinhe o numero entre 1 e 20')
# Ask the player to guess 6 times.
for guessesTaken in range(1, 7):
 print('Acho que é.')
 guess = int(input())
 if guess < secretNumber:
  print('É maior')
 elif guess > secretNumber:
  print('é menor')
 else:
  break # This condition is the correct guess!
if guess == secretNumber:
 print('acertou ' + str(guessesTaken) + ' palpite!')
else:
 print('errou ' + str(secretNumber))
```
