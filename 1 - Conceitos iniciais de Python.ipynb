# OPERADORES

- São simbolos especiais que representam operações em matemática.

```
somar = lambda a,b: a + b
subtrair = lambda a,b: a - b
multiplicar = lambda a,b: a * b
dividir = lambda a,b: a / b
dividir_int = lambda a, b: a // b
modulo = lambda a,b: a % b
exponenciacao =lambda a,b: a ** b


operadores = [somar, subtrair, multiplicar, dividir, dividir_int, modulo, exponenciacao]

for x in operadores:
  print(x(3,2))
```

# DICIONÁRIOS

```
#CRIANDO DICIONÁRIOS
dicionario = {'Carros':'Fusca', 
              'Motos':['Fan 160', ['bmw g310', 'Yamaha  MT 09']], 
              1: 'Qualquer'}
```

```
#RETORNANDO ELEMENTOS ITENS
print(dicionario['Carros'])
print(dicionario.get('Motos'))
print(dicionario['Motos'][1][1])
print(dicionario.get(1))
```

```
#ATULIZANDO ITENS
dicionario['Carros'] = 'Onix'
dicionario['Motos'][0], dicionario['Motos'][1][1] = 'Brós 150', 'Falcon'
dicionario.update({1:'Brayan'})
print(dicionario)
```

## Converter lista em dicionário

```
#CONVERTENDO LISTA EM DICIONÁRIO
contatos_lista = [('Yan', '1234-5678'), ('Pedro', '9999-9999'),
                    ('Ana', '8765-4321'), ('Marina', '8877-7788')]
print(f'{type(contatos_lista)} {contatos_lista}')

print('\t')
#UTLIZANDO FUNÇÃO 'dict'
contatos_dicio = dict(contatos_lista)
print(f'{type(contatos_dicio)} {contatos_dicio}')

print('\t')
#IMPRIMINDO CHAVES
print(contatos_dicio.keys())

print('\t')
#IMPRIMINDO VALORES
print(contatos_dicio.values())
```

## Expressões booleanas
- são expressões que podem ser verdadeiras ou falsas

```
print(5 == '5') # false
print(2 == 2.0) # true
print(1 is 2) # false
print(0 is 0) # true
print('a' is not 'b') # true
print(0 is not 0) # false
print (5 >= 5) # true
print(2 != 2) # false
```
