## Calculo de notas escolares - média


##pergunta nome e idade

```python
loop = True
while loop == True:
  nome = input("Qual seu nome? ")
  idade = input(f"Blz {nome}, qual a sua idade? ")
  print(f"Seu nome é {nome.capitalize()} e você tem {idade}.")
  resposta = input('Deseja continuar? (sim;não)')
  if resposta == 'não':
    loop = False
```

##calculo de nota, passou ou não

```python
notas = input('Digite notas separadas por \'-\'')
nota = notas.split('-')
media = (int(nota[0])+int(nota[1])+int(nota[2]))/3
numero = ''

if media % 2 == 0:
  numero = 'pa'
else:
  numero = 'ímpa'

if media < 6:
  print(f'reprovado, voce tirou {media}, que a saber e: {numero.capitalize()}')
elif media <=10:
  print('Parabens vc passou!')
else:
  print('Digite nota valida')
print(type(notas))
```
